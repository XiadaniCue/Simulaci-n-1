#Aproximaciones

$$x= \frac{U^{0.135} - (1-U)^{0.135}}{0.1975}$$

Aproximamos la distribución Normal

```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt


def aprox(n=10):
  lista=[]

  for i in range(n):
    u=rnd.random()
    x=(u**0.135 - (1-u)**0.135)/0.1975
    lista.append(x)
  return lista
```

```python
A=aprox(10000)
```

```python
#Calculamos la media
mu=np.mean(A)
print(mu)
```

```python
#calculamos la varianza
sigma2=np.var(A)
print(sigma2)
```

```python
#El histograma nos ordena nuestros números por categorías
plt.hist(A,bins=50, color="pink")
plt.title('Funcion de Densidad')
```

```python
#Hacemos nuestra gráfica acumulativa

def cdf(Z):
  n=len(Z)
  x=np.sort(Z)
  y=np.arange(0,n)/n

  return x,y
```

```python
X,Y=cdf(A)
```

```python
plt.step(X,Y, color="blue")
plt.title('Funcion Acumulativa')
plt.axhline(1, linestyle="--", color="orange")
```
