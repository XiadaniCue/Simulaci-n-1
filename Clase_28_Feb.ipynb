```python
from os import major
#Semilla
x0=1
# divisor
m=64
#Constantes
a=13
c=0
suma=0
print ('N | x0 | módulo | xn | xn/m ')
for n in range(0,20):
  x=(a*x0)+c
  xmod= x%m
  div=xmod/m
  print (n, ',', x0, ',',  x, ',', xmod, ',', div)
  x0=xmod
  suma=x0+suma
  media=suma/20
print(media)
#m es quien define la longitud
```
