#**Transformada Inversa**
*Teorema:* Sea U una variable unifome en (0,1). Para cualquier función de dirtibución continua F unvertible, la variable aleatoria x, es definida como
$$x=F^{-1}(u)$$
tiene la distibución F.


**Función Exponencial**

Sea
$$F(x)=1- e^{-λx}$$
invesrtimos nuestra función utilizando nuestro Teorema
$$F(x)=U$$
$$1- e^{-λx}=U$$
$$e^{-λx}=U-1$$
$$ln^{-e^{-λx}}=ln(U-1)$$
$$λx=ln(U-1)$$

$$x=\frac{ln(U-1)}{λ}$$


Nuestro codigo nos queda:



1.   Generar $U\sim U(0,1)$
2.   Si $x=\frac{ln(U-1)}{λ}$
3.   Salir x




```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def expo(landa=2,n=10):
  L=-1/landa
  lista=[]

  for i in range(n):
    u=rnd.random()
    x=L*np.log(u)
    lista.append(x)
  return lista
```

```python
expo()
```

```python
G=expo(0.5,100)
```

```python
#Calculamos la media
mu=np.mean(G)
print(mu)
```

```python
#calculamos la varianza
sigma2=np.var(G)
print(sigma2)
```

```python
#El histograma nos ordena nuestros números por categorías
plt.hist(G,bins=50, color="purple")
plt.title('Gráfica de Densidad')
```

```python
#Se gráfica la secuencia de los números

plt.plot(G)
```

```python
#Hacemos nuestra gráfica acumulativa

def cdf(m):
  n=len(m)
  x=np.sort(m)
  y=np.arange(0,n)/n

  return x,y
```

```python
X,Y=cdf(G)
```

```python
#Calculamos la mediana
med=np.median(G)
print(med)
```

```python
plt.step(X,Y)
plt.axhline(1.0, color="pink", linestyle="--")
plt.title('Función Acumulada')
```
